유무임 승차 비율 분석
1. 데이터를 읽어온다
2. 모든 역의 데이터를 바탕으로 각 역의 비율을 계산
3. 비율이 가장 높은 역을 찾는다(max)
4. 비율이 가장 높은 역이 어디인지, 그 비율이 얼마인지 출력한다.

```python
with open('subwayfee.csv','r',encoding='utf-8') as f:
    data = csv.reader(f)
    next(data)
    mx=0
    rate=0
    for row in data:
        for i in range(4,8):
            row[i]=int("".join(row[i].split(',')))
        if row[6]!=0:
            rate = row[4]/row[6]
            if rate>mx:
                mx=rate
                print(row,round(rate,2))
```

실습 #1: 서울의 일교차
- 기온 공공 데이터를 이용하여 서울에서 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고기온이 가장 높은 날을 찾아 출력하시오.

```python
import csv
from collections import defaultdict
with open('climate_21.csv','r',encoding='cp949') as f:
    data = csv.reader(f)
    min_data = 999.0
    min_date=""
    max_data = -999.0
    max_date=""
    diff_data = 0
    diff_date=""
    next(data)
    for row in data:
        date = row[0]
        min = float(row[3])
        max = float(row[4])
        diff = abs(min-max)
        if(min_data>min):
            min_data = min
            min_date = date
        if(max_data<max):
            max_data = max
            max_date = date
        if(diff_data<diff):
            diff_data = diff
            diff_date = date
    print("일교차가 가장 큰 날:",diff_date,"온도:",diff_data)
    print("최저기온이 가장 낮은 날:",min_date,"온도:",min_data)
    print("최고기온이 가장 높은 날:",max_date,"온도:",max_data)
        
    
```

실습 #2: "지하철 노선별 역별 이용현황" 데이터를 이용하여 노선별로 가장 이용객이 많은 3개역과 가장 적은 3개역을 출력하시오.

```python
import csv
from collections import defaultdict
with open('21_12_subway.csv','r',encoding='utf-8') as f:
    data = csv.reader(f)
    next(data)
    d = defaultdict(list)
    for row in data:
        in_user = int(row[4].replace(',',''))
        out_user = int(row[5].replace(',',''))
        user=in_user+out_user
        station_name = row[3]
        line_data = [station_name,user]
        line = row[1]
        d[line].append(line_data)
    for k in d:
        d[k].sort(key=lambda x:x[1])
        print("노선명:",k)
        for i in range(3):
            print("최저 이용자 #",i+1,"역명:",d[k][i][0],"이용자수:",d[k][i][1])
        for i in range(3):
            print("최대 이용자 #",i+1,"역명:",d[k][-(i+1)][0],"이용자수:",d[k][-(i+1)][1])
```
